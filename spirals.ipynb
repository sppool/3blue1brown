### [Why do prime numbers make these spirals?](https://www.youtube.com/watch?v=EK32jo7i5LQ)

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
nn = 5000
n = nn + 300
arr = np.arange(n)
# arr = arr * (arr % 11 != 0).astype(np.int) * (arr % 4 != 0).astype(np.int)
siz = 25
fig = plt.figure(figsize=(siz, siz))
ax = fig.add_subplot(projection='polar')
ax.set_rlim(0, nn)
ax.set_rticks([])  # Less radial ticks
# ax.plot(arr, arr, 'ro', ms=5)
# ax.plot(arr, arr, lw=0.75, alpha=0.75)
ax.scatter(arr, arr, c=np.minimum(arr, nn), s=0.01*arr, cmap=matplotlib.cm.hsv_r)
plt.grid(False)
```
